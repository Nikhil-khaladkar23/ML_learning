```python
import pandas as pd
import numpy as np
from sklearn.metrics import *
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df= pd.read_csv("C:\\Users\\dhsup\\Desktop\\assingment\\Machine Learning\\Datasets\\Datasets\\Boston.csv")
df.head()
```

```python
df_dum = pd.get_dummies(df)
```

```python
x=df_dum.drop("medv",axis = 1)
y=df_dum["medv"]
```

```python
#spliting data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=24)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
# Polynomial for degree 2
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(x)
print(poly.get_feature_names_out())
```

```python
#spliting the data 
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.3, random_state=24)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
print(lr.coef_)
print(lr.intercept_)
```

```python
y_pred = lr.predict(X_test)
print('r2:',r2_score(y_test, y_pred))
```

```python
# Polynomial for degree 3
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(x)
print(poly.get_feature_names_out())
```

```python
#spliting the data 
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.3, random_state=24)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
print(lr.coef_)
print(lr.intercept_)
```

```python
y_pred = lr.predict(X_test)
print('r2:',r2_score(y_test, y_pred))
```
