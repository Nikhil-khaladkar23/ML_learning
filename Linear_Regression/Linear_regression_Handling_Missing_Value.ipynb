```python
import pandas as pd
import numpy as np
from sklearn.metrics import *
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
df= pd.read_csv("C:\\Users\\dhsup\\Downloads\\ChemicalProcess.csv")
df.head()
```

```python
df_dum = pd.get_dummies(df)
```

```python
x=df_dum.drop("Yield",axis = 1)
y=df_dum["Yield"]
```

```python
#spliting data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=24)
```

```python
imp = SimpleImputer(strategy="mean").set_output(transform="pandas")
## train set opreation
x_trn_imp=imp.fit_transform(x_train)
```

```python
model = LinearRegression()
model.fit(x_trn_imp,y_train)
```

```python
x_test_imp = imp.transform(x_test)
y_pred=model.predict(x_test_imp)
```
